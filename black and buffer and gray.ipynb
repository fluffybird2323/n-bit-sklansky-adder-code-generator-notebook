```python
def printcell(s,i,j,k,a,b):
    if(s=="black"):
        printblack(i,j,k,a,b)
    elif(s=="buffer"):
        printbuffer(i,j,a,b)
    else :
        printgrey(i,j,k,a,b)
```

```python
def printverilog(n):
    #initialiseverilog(n)
    print_wires(n)
    for b in range(1+floor(log(n)/log(2))):   
        for a in range(n):
            s = findcell(b,a)
            i,j,k = findijk(a,b)
            printcell(s,i,j,k,a,b)
```

```python
def findcell(a,b):
    s = ""
    if((b//(2**(a-1)))%2==0):
        s = "buffer"
    elif((b//(2**(a-1)))==1):
        s = "grey"
    else:
        s = "black"
    return s
def findijk(a,b):
    j=0
    j = (2*a)*(b//(2**a))
    i = b
    a = a-1
    k = (2*a)*(b//(2**a))
    return i,j,k;
```
