```python
def printverilog(n):
    initializeGP()
    initialiseverilog(n)
    print_wires(n)
    assignGP(n)
    print("assign G_1_0 = 0;")
    for a in range(1,1+ceil(log(n)/log(2))):
        for b in range(n):
            s = findcell(a,b)
            i,j,k = findijk(a,b)
            printcell(s,i,j,k,a,b)
    computeS(n)        
    print("endmodule")
```

```python
def findcell(a,b):
    s = ""
    if((b//(2**(a-1)))%2==0):
        s = "buffer"
    elif((b//(2**(a-1)))==1):
        s = "grey"
    else:
        s = "black"
    return s
def findijk(a,b):
    j=0
    j = (2*a)*(b//(2**a))
    i = b
    a = a-1
    k = (2*a)*(b//(2**a))
    if(a==0):
        k = i

    return i,j,k;
```

```python
def printmodule(n):
    argin=""
    argout=""
    
    for i in range(n):
        argin+="A"+str(i)+","
    for i in range(n):
        argin+="B"+str(i)+","
    for i in range(n):
        argout+="S"+str(i)+","
    argout = argout[: -1]
    print("module sklansky("+""+argin+argout+")"+";")
    print("input "+""+ argin[: -1]+";")
    print("output "+""+ argout+";")
```

```python
def initializeGP():
    print("module initializeGP(A,B,G,P);")
    print("input A,B;")
    print("output G,P;")    
    print("assign G = A^B;")
    print("assign P = A&B;")   
    print("endmodule")
    print()
```

```python
def assignGP(n):
    for i in range(n):
        print("initializeGP(A"+str(i)+",B"+str(i)+",G"+str(i)+"_"+str(i)+",P"+str(i)+"_"+str(i)+");")
    print()
```

```python
def computeS(n):
    print("assign S"+str(0)+" = G_1_0"+"^"+"P"+str(0)+"_"+str(0)+";")
    for i in range(1,n):
        print("assign S"+str(i)+" = G"+str(i-1)+"_0^"+"P"+str(i)+"_"+str(i)+";")
```
